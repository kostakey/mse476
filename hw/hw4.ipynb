```python
import numpy as np
import matplotlib.pyplot as plt

# --- Given constants ---
K1 = 10**0.9   # 7.943
K2 = 10**0.65  # 4.467
K3 = 10**-1.5  # 0.0316
```

```python
# --- Temperature range (°C) ---
T = np.linspace(526, 726, 200)
DT = 727 - T  # undercooling (ΔT)

# --- All units in µm and µm/s ---
# 1 mm = 1000 µm

# Eq. (18.3): ΔT = K1 * sqrt(V)
# Convert velocity from mm/s → µm/s by multiplying by 1000
V_mm = (DT / K1)**2              # in mm/s
V = V_mm * 1000                  # in µm/s

# Eq. (18.4): λ = K2 / sqrt(V_mm)
# K2 constant is in µm·(mm/s)^(1/2)
l = K2 / np.sqrt(V_mm)           # in µm

# Eq. (18.5): 1/λ = K3 * ΔT
l_alt = 1 / (K3 * DT)            # in µm
```

```python
# --- Plot (a): T vs log(V) ---
plt.figure(figsize=(6,4))
plt.plot(T, np.log10(V), 'b')
plt.xlabel('Temperature (°C)')
plt.ylabel('log₁₀(V)  [µm/s]')
plt.title('(a) Temperature vs Growth Velocity (µm/s)')
# plt.grid(True)
plt.gca().invert_xaxis()
plt.show()

# --- Plot (b): T vs log(λ) ---
plt.figure(figsize=(6,4))
plt.plot(T, np.log10(l), 'r')
plt.xlabel('Temperature (°C)')
plt.ylabel('log₁₀(λ)  [µm]')
plt.title('(b) Temperature vs Interlamellar Spacing (µm)')
# plt.grid(True)
plt.gca().invert_xaxis()
plt.show()

# --- Plot (c): λ vs V (log-log scale) ---
plt.figure(figsize=(6,4))
plt.plot(V, l, 'g')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Growth Velocity, V [µm/s]')
plt.ylabel('Interlamellar Spacing, λ [µm]')
plt.title('(c) Interlamellar Spacing vs Growth Velocity')
# plt.grid(True, which='both')
plt.show()

# --- Optional consistency check ---
print("Check K3 consistency:")
print(f"K3 (given) = {K3:.4f}")
print(f"1 / (K1 * K2) = {1 / (K1 * K2):.4f}")
```
